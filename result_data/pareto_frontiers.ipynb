```python
import csv
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot_pareto_frontier(Xs, Ys, names, maxX=True, maxY=True):
    '''Pareto frontier selection process'''
    sorted_list = sorted([[Xs[i], Ys[i], names[i]] for i in range(len(Xs))], reverse=maxY, key=lambda x: x[:1])
    pareto_front = [sorted_list[0]]
    for pair in sorted_list[1:]:
        if maxY:
            if pair[1] >= pareto_front[-1][1]:
                pareto_front.append(pair)
        else:
            if pair[1] <= pareto_front[-1][1]:
                pareto_front.append(pair)
    
    '''Plotting process'''
    plt.rcParams["figure.figsize"] = (25,30)
    plt.rcParams.update({'font.size': 22})
    plt.scatter(Xs,Ys)
    pf_X = [pair[0] for pair in pareto_front]
    pf_Y = [pair[1] for pair in pareto_front]
    plt.plot(pf_X, pf_Y, marker = '.', markersize=18)
    plt.axhline(y=30, color='r', linestyle="dashed")
    plt.xlabel("IoU score")
    plt.ylabel("frames per second")
    
    for x, y, name in pareto_front:
        label = name

        plt.annotate(label, # this is the value which we want to label (text)
                     (x,y), # x and y is the points location where we have to label
                     textcoords="offset points",
                     xytext=(0,10), # this for the distance between the points
                     # and the text label
                     ha='center',
                     arrowprops=dict(arrowstyle="->", color='red')
                     )
    plt.show()
```

```python
file = open("stage1.csv")
reader = csv.reader(file)
ious = []
fps = []
names = []
for i, row in enumerate(reader):
    if i==0:
        continue
    ious.append(float(row[7]))
    fps.append(float(row[6]))
    names.append(str(row[0]))

plot_pareto_frontier(ious, fps, names)
```
